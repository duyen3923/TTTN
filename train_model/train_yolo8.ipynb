Train with yolov8

```python
from ultralytics import YOLO

# Load model
model = YOLO('yolov8m.pt')

# Train with augmentation
results = model.train(
    data='data.yaml',
    epochs=100,
    imgsz=640,
    batch=24,
    name='model_train_aug',
    save=True,
    device=[0, 1],
    augment=True,  # bật tăng cường dữ liệu
    hsv_h=0.015,   # thay đổi màu sắc (Hue)
    hsv_s=0.7,     # thay đổi độ bão hòa (Saturation)
    hsv_v=0.4,     # thay đổi độ sáng (Value)
    degrees=10,    # xoay ảnh
    translate=0.1, # tịnh tiến
    scale=0.5,     # thu phóng
    shear=0.1,     # biến dạng nghiêng
    flipud=0.0,    # lật dọc
    fliplr=0.5,    # lật ngang
    mosaic=1.0,    # mosaic augmentation
    mixup=0.2      # mixup augmentation
)
```
