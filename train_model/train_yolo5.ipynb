```python
from ultralytics import YOLO
import pandas as pd

model = YOLO('yolov5m.pt')
results = model.train(
    data='data.yaml',
    epochs=80,
    imgsz=640,
    batch=24,
    name='model_train',
    save=True,
    device=[0,1],
    augment=True,  # báº­t tÄƒng cÆ°á»ng dá»¯ liá»‡u
    hsv_h=0.015,   # thay Ä‘á»•i mÃ u sáº¯c (Hue)
    hsv_s=0.7,     # thay Ä‘á»•i Ä‘á»™ bÃ£o hÃ²a (Saturation)
    hsv_v=0.4,     # thay Ä‘á»•i Ä‘á»™ sÃ¡ng (Value)
    degrees=10,    # xoay áº£nh
    translate=0.1, # tá»‹nh tiáº¿n
    scale=0.5,     # thu phÃ³ng
    shear=0.1,     # biáº¿n dáº¡ng nghiÃªng
    flipud=0.0,    # láº­t dá»c
    fliplr=0.5,    # láº­t ngang
    mosaic=1.0,    # mosaic augmentation
    mixup=0.2      # mixup augmentation
)


pd.DataFrame(results).to_csv("model_train_results.csv")
```
