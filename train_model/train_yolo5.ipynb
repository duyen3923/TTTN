```python
yaml_content = """
path: /kaggle/input/d/duynngo/vocdev/SODA/images/
train: train
val: val

nc: 2
names: ['person','helmet']
"""

with open("data.yaml", "w") as f:
    f.write(yaml_content)
print("done")
```

```python
from ultralytics import YOLO
import pandas as pd

model = YOLO('yolov5m.pt')
results = model.train(
    data='data.yaml',
    epochs=80,
    imgsz=640,
    batch=24,
    name='model_train',
    save=True,
    device=[0,1],
    augment=True,  # bật tăng cường dữ liệu
    hsv_h=0.015,   # thay đổi màu sắc (Hue)
    hsv_s=0.7,     # thay đổi độ bão hòa (Saturation)
    hsv_v=0.4,     # thay đổi độ sáng (Value)
    degrees=10,    # xoay ảnh
    translate=0.1, # tịnh tiến
    scale=0.5,     # thu phóng
    shear=0.1,     # biến dạng nghiêng
    flipud=0.0,    # lật dọc
    fliplr=0.5,    # lật ngang
    mosaic=1.0,    # mosaic augmentation
    mixup=0.2      # mixup augmentation
)


pd.DataFrame(results).to_csv("model_train_results.csv")
```
